```python
from parrot import Parrot
import torch
import warnings
warnings.filterwarnings("ignore")
```

```python
import nltk
```

```python
nltk.download('punkt')
```

```python
def random_state(seed):
  torch.manual_seed(seed)
  if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

random_state(1234)
```

```python
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5", use_gpu=False)
```

```python
text = "What is the best movie of 2022?"
```

```python
phrases = nltk.tokenize.sent_tokenize(text)
```

```python
phrases
```

```python
para = []
```

```python
for phrase in phrases:
  sentence = []
  para_phrases = parrot.augment(input_phrase=phrase)
  for para_phrase in para_phrases:
   sentence.append(para_phrase)
  para.append(sentence)
```

```python
import random
```

```python
paraphrased_response = []
```

```python
for i in range(3):
  text = " " 
  for j in range(len(phrases)):
    text = text + para[j][random.randint(0,len(para[j])-1)][0] + ". "
  paraphrased_response.append(text)
```

```python
for i in range(3):
  print("Suggestion ",i+1,": ", paraphrased_response[i])
```
